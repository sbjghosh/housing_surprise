```python
def meanplot(a):
    print('First figure: Varition of avarage sale price with', a)
    df=data[[a,'SalePrice']]
    df=df.sort_values([a])
    df = df.groupby([a])[['SalePrice']].mean().reset_index()
    df = df.dropna()
    sns.barplot(x=a,y='SalePrice',data=df)
    
    
    
def countplot(a):
    print('Second Figure: Varition of count of sales with', a)
    df=data[[a,'SalePrice']]
    df=df.sort_values([a])
    df = df.groupby([a])[['SalePrice']].count().reset_index()
    df = df.dropna()
    sns.barplot(x=a,y='SalePrice',data=df)

def catplot(a):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    meanplot(a)
    plt.subplot(1,2,2)
    countplot(a)
    
#sns.barplot(x='Neighborhood',y='SalePrice',data=df)
#plt.plot()
catplot('LotConfig')
```

```python
def diplot(a):
    print('First figure: Distribution of', a)
    df=data[[a,'SalePrice']]
    df=df.dropna()
    sns.distplot(df[a])
    
    
    
def scaplot(a):
    print('Second Figure: Varition sales with', a)
    df=data[[a,'SalePrice']]
    df = df.dropna()
    plt.plot(df[a],df['SalePrice'],'o')

def numplot(a):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    diplot(a)
    plt.subplot(1,2,2)
    scaplot(a)
    
#sns.barplot(x='Neighborhood',y='SalePrice',data=df)
#plt.plot()
numplot('LotArea')
```
